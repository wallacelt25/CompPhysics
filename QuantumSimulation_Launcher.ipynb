## Step 1: Install Required Dependencies

The following cell will install all necessary Python packages for the quantum simulation.

```python
import sys
import subprocess
import importlib

# List of required packages
required_packages = [
    'numpy>=1.21.0',
    'scipy>=1.7.0', 
    'matplotlib>=3.4.0',
    'numba>=0.54.0',
    'pillow>=8.0.0',
    'pandas>=1.3.0',
    'seaborn>=0.11.0'
]

def install_package(package):
    """Install a package using pip"""
    try:
        subprocess.check_call([sys.executable, '-m', 'pip', 'install', package])
        print(f"âœ… Successfully installed {package}")
    except subprocess.CalledProcessError as e:
        print(f"âŒ Failed to install {package}: {e}")
        return False
    return True

def check_package(package_name):
    """Check if a package is already installed"""
    try:
        importlib.import_module(package_name)
        return True
    except ImportError:
        return False

print("ğŸ”§ Installing required packages for Quantum Wave Packet Simulation...\n")

# Check and install packages
for package in required_packages:
    package_name = package.split('>=')[0].split('==')[0]
    
    if check_package(package_name):
        print(f"âœ… {package_name} is already installed")
    else:
        print(f"ğŸ“¦ Installing {package}...")
        install_package(package)

print("\nğŸ‰ Package installation completed!")
print("\nğŸ“‹ Verifying installations...")

# Verify installations
verification_packages = ['numpy', 'scipy', 'matplotlib', 'numba', 'PIL', 'pandas', 'seaborn']
all_installed = True

for pkg in verification_packages:
    if check_package(pkg):
        print(f"âœ… {pkg} - OK")
    else:
        print(f"âŒ {pkg} - NOT FOUND")
        all_installed = False

if all_installed:
    print("\nğŸ¯ All packages successfully installed and verified!")
else:
    print("\nâš ï¸  Some packages may not be properly installed. Please check the output above.")
```
